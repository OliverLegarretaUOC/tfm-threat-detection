```python
# =========================
# Generar no_gun_test.txt
# Recorre No_Gun/ y escribe una línea por vídeo encontrado
# =========================
from pathlib import Path

# -------- CONFIG --------
NO_GUN_ROOT = "/content/drive/MyDrive/TFM/datasets/videos/Gun_Action_Recognition_Dataset/No_Gun"
OUT_TXT     = "/content/drive/MyDrive/TFM/datasets/videos/Gun_Action_Recognition_Dataset/splits/no_gun_test.txt"
# ------------------------

no_gun_root = Path(NO_GUN_ROOT)
assert no_gun_root.exists(), f"No existe: {NO_GUN_ROOT}"

video_paths = sorted(no_gun_root.rglob("video.mp4"))
assert len(video_paths) > 0, "No se encontraron vídeos en No_Gun/"

out_txt = Path(OUT_TXT)
out_txt.parent.mkdir(parents=True, exist_ok=True)
out_txt.write_text("\n".join(str(p) for p in video_paths) + "\n")

print(f"✅ Generado: {OUT_TXT}")
print(f"   Total vídeos negativos: {len(video_paths)}")

# Desglose por categoría (N1, N2, ...)
from collections import Counter
cats = Counter(p.parent.name.split("_")[0] for p in video_paths)
print("\nDesglose por categoría:")
for cat in sorted(cats, key=lambda x: int(x[1:])):
    print(f"  {cat}: {cats[cat]} vídeos")
```
